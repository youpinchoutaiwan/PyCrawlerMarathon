# 從104人力銀行網站爬取求職公司資訊
進入104人力銀行網站選擇"地區"及"產業"然後按下"找公司"按鈕。開啟新網頁後，觀察新的網址內容，把網址複製下來。

```python
import time
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
```

```python
url="https://www.104.com.tw/cust/list/index/?page=1&order=1&mode=s&jobsource=checkc&area=6001012000&indcat=1005000000"
#爬取搜尋結果
browser=webdriver.Chrome()
browser.get(url)
while True:
    A=browser.page_source
    soup=BeautifulSoup(A,"html5lib")
    companys=soup.find_all("article",{"class":"items"})
    
    #擷取公司名稱、簡介
    for b in companys:
        #內容太多，打開一個檔案，把資料寫入
#         with open("companylist.txt","w",encoding="utf-8") as fh:
            name=b.find("h1").a.string
            print(name+"\n")
            intro=b.find("p",{"class":"desc"})
            print(intro.string+"\n")
            print("----------------------------------------------------")
            
        
    if soup.find("a",{"class":"page-next disabled"}) !=None:
                break

    browser.find_element_by_link_text("下一頁").click()
```
