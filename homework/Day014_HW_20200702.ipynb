```python
#確認上映中電影數量
import requests
from bs4 import BeautifulSoup
url="https://movies.yahoo.com.tw/movie_intheaters.html"
res=requests.get(url)
soup=BeautifulSoup(res.text,"html5lib")
number=soup.find("div",{"class":"release_box"})
print(number.p.string)
```

```python
#爬取所有上映中的電影資料：電影名稱、期待度、滿意度、簡介等等。(滿意度跑步出來)
max_page=5
for page_number in range(1,max_page+1,1):
    url="https://movies.yahoo.com.tw/movie_intheaters.html"
    payload={"page":str(page_number)}
    res=requests.get(url,params=payload)
    res.encoding="utf-8"
    soup=BeautifulSoup(res.text,"html5lib")
    
    movie_list=soup.find("ul",{"class":"release_list"}).find_all("li")
    for p in movie_list:
        #電影名稱
        movie_name = p.find("div", {"class":"release_movie_name"})
        print("電影名稱:",movie_name.a.string)
        
        #期待度
        level_box=movie_name.find("dl",{"class":"levelbox"})
        exception=level_box.find("div",{"class":"leveltext"})
        print("期待度:",exception.span.string)
        
        #簡介
        intro=p.find("div",{"class":"release_text"})
        print("簡介:",intro.span.string)
        print("-------------------------------")
```

```python
#爬取上映電影的ID
import re
A=soup.find("select",{"class":"movie_name"})
ids=A.find_all("option",{"data-name":re.compile(".*")})
for i in ids:
    print("電影: %s, ID:%s" % (i["data-name"],i["value"]))
```
