```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
url="https://www.ettoday.net/news/news-list.htm"
res=requests.get(url)
soup=BeautifulSoup(res.text,"html5lib")
A=soup.find("div",{"class":"part_list_2"})
```

```python
#取出所有標題
titles=A.find_all("a")
for title in titles:
    print(title.text)
```

```python
timelist=[]
typelist=[]
namelist=[]
h3s=A.find_all("h3")
for h3 in h3s:
    timelist.append(h3.find("span").text)
    typelist.append(h3.find("em").text)
    namelist.append(h3.find("a").text)
dic=dict(類別=typelist,時間=timelist,標題=namelist)
pd.set_option("display.max_row",None)
items=pd.DataFrame(dic)
print(items)
print("===================================")
items.sort_values(["類別"])
```

```python
#抓出特定標題
B=items["類別"].str.contains("社會")
print(items[B])
```

```python
#哪個類別文章最多
df2=pd.DataFrame(items.groupby("類別").size(), columns=["文章數量"])
df2["排名"]=df2.rank(ascending=False)
df2.sort_values(by=["排名"])
```
