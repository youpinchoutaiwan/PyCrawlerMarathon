# Day31 尋找一個網站是有作 Header 檢查的（Hint: 有加跟沒有加回傳結果不一樣）

```python
import requests
res=requests.get("https://zhihu.com")
res.text
```

```python
headers={"user-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.116 Safari/537.36"}
res=requests.get("https://zhihu.com",headers=headers)
res.text
```
