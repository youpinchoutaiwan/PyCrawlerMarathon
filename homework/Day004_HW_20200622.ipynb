```python
#json讀取的檔案以key-value的字典型態存取
import json
import urllib.request as requests
s="https://data.taipei/api/v1/dataset/296acfa2-5d93-4706-ad58-e83cc951863c?scope=resourceAquire"
with requests.urlopen(s) as response:
    fh=json.load(response)
    print(fh)
```

```python
#只取出公司名稱
clist=fh["result"]["results"]
for company in clist:
    print(company["公司名稱"])
print("----------------------")
for company in clist:    
    print(company["\ufeff統編"])
```
