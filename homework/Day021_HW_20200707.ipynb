# 取出今天所有的新聞

```python
from selenium import webdriver
from bs4 import BeautifulSoup

#打開瀏覽器
browser=webdriver.Chrome(executable_path="chromedriver")
browser.get("https://www.ettoday.net/news/news-list-2020-7-7-0.htm")

#每隔兩分鐘自動下滑
import time
for i in range(60):
    time.sleep(1)
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")

#取得資料交由BS解析
A=browser.page_source
soup=BeautifulSoup(A,"html5lib")

#爬出新聞時間、名稱
news={}
for a in soup.find("div",{"class":"part_list_2"}).find_all("h3"):
    news.update({a.find(class_="date").text:a.find_all("a")[-1].text})
browser.close()
#print(news)
```

```python
import datetime
for key, value in news.items():
    d=datetime.datetime.strptime(key, "%Y/%m/%d %H:%M")
    if (d.year==2020) and (d.month==7) and (d.day==7):
        print(key, value)
```

# 取出現在時間2HR內的新聞

```python
for key, value in news.items():
    d=datetime.datetime.strptime(key, "%Y/%m/%d %H:%M")
    if (d.year==2020) and (d.month==7) and (d.day==7) and (d.hour==18) or(d.hour==17):
        print(key, value)
```

# 根據範例，取出三天前下午三點到五點的新聞

```python
browser=webdriver.Chrome(executable_path="chromedriver")
browser.get("https://www.ettoday.net/news/news-list-2020-7-4-0.htm")

import time
for i in range(60):
    time.sleep(1)
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    
B=browser.page_source
soup1=BeautifulSoup(B,"html5lib")

news={}
for d in soup1.find(class_="part_list_2").find_all("h3"):
    news.update({d.find(class_="date").text : d.find_all("a")[-1].text})
    
browser.close()

for key, value in news.items():
    d=datetime.datetime.strptime(key, "%Y/%m/%d %H:%M")
    if (d.year==2020) and (d.month==7) and (d.day==4)and ((d.hour==15) or (d.hour==16)):
        print(key, value)
```
