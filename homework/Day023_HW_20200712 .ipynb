# Ettoday動態爬蟲練習從東森新聞雲網站中，爬取新聞摘要及詳細內容。

```python
import requests
import time
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
```

# 以國際新聞為例，先連結到生活新聞首頁，用selenium模擬往下拉網頁

```python
url="https://www.ettoday.net/news/focus/%E8%B2%A1%E7%B6%93/"
browser=webdriver.Chrome()

browser.get(url)

#模擬往下拉網頁
last_height=browser.execute_script("return document.body.scrollHeight")
while True:
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")  #scroll down to bottom
    time.sleep(1)  #wait time to load
    new_height=browser.execute_script("return document.body.scrollHeight")  #calculate new height
    if new_height==last_height:
        print("到達頁面底端")
        break
    print("網頁更新中")
    last_height=new_height
    
    
    
```

# 到達此步驟，所有新聞都已在頁面上，可以開始爬取

```python
#建立一個函數，其功能為 (1)連結外部連結 (2)爬取其內容
def getdata(link):
    res=requests.get(link)
    res.encoding="utf-8"
    soup=BeautifulSoup(res.text,"html5lib")
    content=soup.find("div",{"class":"story"}).find_all("p")
    for a in content:
        if ((a.string) is not None):
            print(a.string)
```

# 解析HTML並萃取新聞摘要，若有外部的連結，再連到外部連結並把詳細新聞內容抓取下來。

```python
#爬取網頁新聞摘要
A=browser.page_source
soup=BeautifulSoup(A,"html5lib")
all_news=soup.find("div",{"class":"block block_1 infinite_scroll"})
titles=all_news.find_all("div",{"class":"piece clearfix"})
for i,item in enumerate(titles):
    body=item.find("h3")
    print("\n[%d] %s\n" % (i, body.a.string))
   
    external_link="https://www.ettoday.net"+body.a ["href"]
    getdata(external_link)
    print("-----------------------------------")
```
