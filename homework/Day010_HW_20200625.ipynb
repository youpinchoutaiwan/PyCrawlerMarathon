```python
import requests
from bs4 import BeautifulSoup
res=requests.get("https://google.com")
soup=BeautifulSoup(res.text)
print(soup)
```

```python
from grab import Grab
from pyquery import PyQuery as pq
g=Grab()
resp=g.go("https://google.com")
doc=pq(resp.body)
print(doc)
```
