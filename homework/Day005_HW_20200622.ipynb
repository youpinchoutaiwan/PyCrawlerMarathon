```python
import urllib.request as req
import json
pop="https://www.dcard.tw/_api/forums/pet/posts?popular=true"
pop1=req.Request(pop,headers={
    "user-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.106 Safari/537.36"
})
with req.urlopen(pop1) as response:
    A=json.load(response)
    print(A)
```

```python
#熱門貼文總共幾則，有哪些標籤
with req.urlopen(pop1)as response:
    fh=json.load(response)
    print(len(fh))
    print(fh[0].keys())
```

```python
#取出每筆資料「標題」、「貼文時間」、「留言人數」、「按讚人數」
with req.urlopen(pop1) as response:
    fh=json.load(response)
    for c in fh:
        print(c["updatedAt"],c["title"],"留言人數:",c["commentCount"],"按讚人數:",c["likeCount"])
```

```python
#計算熱門「平均留言人數」與「平均按讚人數」
with req.urlopen(pop1) as response:
    fh=json.load(response)
    comment=0
    like=0
    for data in fh:
        comment+=data["commentCount"]
        like+=data["likeCount"]
print(comment/len(fh))
print(like/len(fh))     
```

```python
#計算冷門「平均留言人數」與「平均按讚人數」
unpop="https://www.dcard.tw/_api/forums/pet/posts?popular=false"
unpop1=req.Request(unpop,headers={
    "user-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.106 Safari/537.36"
})
with req.urlopen(unpop1) as response:
    A=json.load(response)
    comment=0
    like=0
    for c in A:
        comment+=c["commentCount"]
        like+=c["likeCount"]
print(comment/len(A))
print(like/len(A))
```
